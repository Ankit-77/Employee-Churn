# KNN

```python
from sklearn.preprocessing import StandardScaler
print(list(main_df.columns))
```

```python
num_features = list(main_df.columns)
for x in ['salary','department','left']:
    num_features.remove(x)
    
sc = StandardScaler()

X_train_std=sc.fit_transform(X_train[num_features])
X_test_std=sc.transform(X_test[num_features])
```

```python
from sklearn.neighbors import KNeighborsClassifier

k_range=range(1,26)
scores={}
scores_list=[]
bestk=0
h_score = 0

for i in range(1,26):
    clf = KNeighborsClassifier(n_neighbors=i).fit(X_train_std,y_train)
    pred = clf.predict(X_test_std)
    scores[i]=accuracy_score(y_test,pred)
    if scores[i]>h_score:
        h_score = scores[i]
        bestk = i
    scores_list.append(scores[i]*100)
#print(scores_list)
print('Best k is {} with score : {}'.format(bestk,h_score*100))
```

```python
knn=KNeighborsClassifier(n_neighbors=bestk)
knn.fit(X_train_std,y_train)
```

```python
prediction_knn=knn.predict(X_test_std)
```

```python
prediction_knn
```

```python
accuracy_knn=accuracy_score(y_test,prediction_knn)*100
print('Accuracy score : ',accuracy_knn)
scores_dict['KNeighborsClassifier'] = accuracy_score(y_test,prediction_knn)*100
```

```python
cf_matrix_knn = confusion_matrix(y_test, prediction_knn)
group_names = ['True Neg','False Pos','False Neg','True Pos']
group_counts = ["{0:0.0f}".format(value) for value in
                cf_matrix_knn.flatten()]
group_percentages = ["{0:.2%}".format(value) for value in
                     cf_matrix_knn.flatten()/np.sum(cf_matrix_knn)]
labels = [f"{v1}\n{v2}\n{v3}" for v1, v2, v3 in
          zip(group_names,group_counts,group_percentages)]
labels = np.asarray(labels).reshape(2,2)
sns.heatmap(cf_matrix_knn, annot=labels, fmt='', cmap='binary')
```

```python
print(classification_report(y_test,prediction_knn))
```

```python
fig = px.line(x = k_range,y = scores_list, markers = True, title='Best k value')
fig.update_layout(xaxis_title="k value", yaxis_title="Accuracy")
#fig.show()
fig.show("png")
```

```python
from sklearn.metrics import roc_curve, auc
false_positive_rate_knn, true_positive_rate_knn, thresholds = roc_curve(y_test, prediction_knn)
roc_auc_knn = auc(false_positive_rate_knn, true_positive_rate_knn)
roc_auc_knn
```

# ANN

```python
X=main_df.drop(['left'],axis=1)
y = main_df.left
```

```python
X
```

```python
y
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.2,random_state=5)
```

```python
X_train.shape
```

```python
X_test.shape
```

```python
len(X_train.columns)
```

```python
from sklearn.preprocessing import StandardScaler
num_features = list(main_df.columns)
for x in ['salary','department','left']:
    num_features.remove(x)

sc = StandardScaler()

X_train_std=sc.fit_transform(X_train[num_features])
X_test_std=sc.transform(X_test[num_features])
```

```python
import tensorflow as tf
from tensorflow import keras
```

```python
ANN_model = tf.keras.models.Sequential([
    tf.keras.layers.Dense(256, activation = 'relu'), # ReLU : rectified linear activation unit
    tf.keras.layers.Dense(128, activation = 'relu'),
    tf.keras.layers.Dense(64, activation = 'relu'),
    tf.keras.layers.Dense(32, activation = 'relu'),
    tf.keras.layers.Dense(16, activation = 'relu'),
    tf.keras.layers.Dense(8, activation = 'relu'),
    tf.keras.layers.Dense(4, activation = 'relu'),
    tf.keras.layers.Dense(1, activation = 'sigmoid'), # for binary its sigmoid for multi-class it is softmax
])
```

```python
from tensorflow.keras.optimizers import Adam

ANN_model.compile(
    optimizer = Adam(),
    loss = 'binary_crossentropy',
    metrics = ['accuracy']
)
```

```python
history = ANN_model.fit(
    X_train_std, y_train,
    validation_data = (X_test_std, y_test),
    epochs = 10,  
    verbose = 1
)
```

```python
ANN_model.evaluate(X_test_std, y_test)
```

```python
acc = history.history['accuracy']
val_acc = history.history['val_accuracy']

plt.figure()
plt.plot(acc, color = 'green', label = 'Training accuracy')
plt.plot(val_acc, color = 'red', label = 'Validation accuracy')
plt.legend()
```

```python
yp = ANN_model.predict(X_test_std)
yp
```

```python
y_pred = []
for element in yp:
    if element > 0.5:
        y_pred.append(1)
    else:
        y_pred.append(0)
```

```python
y_pred[:5]
```

```python
ann_accuracy = accuracy_score(y_test,y_pred)*100
print('Accuracy score : ',ann_accuracy)
scores_dict['ANN'] = ann_accuracy
print(classification_report(y_test,y_pred))
```

```python
cf_matrix_ann = confusion_matrix(y_test, y_pred)
group_names = ['True Neg','False Pos','False Neg','True Pos']
group_counts = ["{0:0.0f}".format(value) for value in
                cf_matrix_ann.flatten()]
group_percentages = ["{0:.2%}".format(value) for value in
                     cf_matrix_ann.flatten()/np.sum(cf_matrix_ann)]
labels = [f"{v1}\n{v2}\n{v3}" for v1, v2, v3 in
          zip(group_names,group_counts,group_percentages)]
labels = np.asarray(labels).reshape(2,2)
sns.heatmap(cf_matrix_ann, annot=labels, fmt='', cmap='binary')
```

```python
from sklearn.metrics import roc_curve, auc
false_positive_rate_ann, true_positive_rate_ann, thresholds = roc_curve(y_test, y_pred)
roc_auc_ann = auc(false_positive_rate_ann, true_positive_rate_ann)
roc_auc_ann
```
